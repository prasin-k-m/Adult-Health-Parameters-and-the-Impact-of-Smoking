```python
def add_bmi_labels(ax, df, column='BMI'):

    total = len(df)
    under = ((df[column] < 18.5).sum() / total) * 100
    normal = ((df[column] >= 18.5) & (df[column] < 24.9)).sum() / total * 100
    over = ((df[column] >= 24.9) & (df[column] < 29.9)).sum() / total * 100
    obese = ((df[column] >= 29.9)).sum() / total * 100

    labels = [ f"U:{under:.1f}%",
               f"N:{normal:.1f}%",
               f"OV:{over:.1f}%",
               f"OB:{obese:.1f}%" ]


    positions = [9, 21, 27, 33]
    
    for pos, label in zip(positions, labels):
        ax.text(pos, ax.get_ylim()[1] * 0.8, label, fontsize=10,
                ha='center', va='center', bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

fig, axs = plt.subplots(2, 2, figsize=(18, 10)) 

#female
sns.kdeplot(df3['BMI'], color='red', fill=True, ax=axs[0,0])
axs[0,0].set_title('BMI Distribution - Female')
axs[0,0].set_xlabel('BMI')
axs[0,0].set_ylabel('Density')

axs[0,0].axvline(18.5, color='green', linestyle='--')
axs[0,0].axvline(24.9, color='orange', linestyle='--')
axs[0,0].axvline(29.9, color='red', linestyle='--')

axs[0,0].axvspan(0, 18.5, color='cyan', alpha=0.15, label='Underweight')
axs[0,0].axvspan(18.5, 24.9, color='green', alpha=0.15, label='Normal')
axs[0,0].axvspan(24.9, 29.9, color='orange', alpha=0.15, label='Overweight')
axs[0,0].axvspan(29.9, df3['BMI'].max(), color='red', alpha=0.15, label='Obese')
axs[0,0].legend()
add_bmi_labels(axs[0,0], df3)

#male
sns.kdeplot(df4['BMI'], color='blue', fill=True, ax=axs[0,1])
axs[0,1].set_title('BMI Distribution - Male')
axs[0,1].set_xlabel('BMI')
axs[0,1].set_ylabel('Density')

axs[0,1].axvline(18.5, color='green', linestyle='--')
axs[0,1].axvline(24.9, color='orange', linestyle='--')
axs[0,1].axvline(29.9, color='red', linestyle='--')

axs[0,1].axvspan(0, 18.5, color='cyan', alpha=0.15, label='Underweight')
axs[0,1].axvspan(18.5, 24.9, color='green', alpha=0.15, label='Normal')
axs[0,1].axvspan(24.9, 29.9, color='orange', alpha=0.15, label='Overweight')
axs[0,1].axvspan(29.9, df4['BMI'].max(), color='red', alpha=0.15, label='Obese')
axs[0,1].legend()
add_bmi_labels(axs[0,1], df4)


#combined
sns.kdeplot(df4['BMI'], color='blue', fill=True, label='Male', ax=axs[1,0])
sns.kdeplot(df3['BMI'], color='red', fill=True, label='Female', ax=axs[1,0])

axs[1,0].set_title('BMI Distribution - Male vs Female')
axs[1,0].set_xlabel('BMI')
axs[1,0].set_ylabel('Density')

axs[1,0].axvline(18.5, color='green', linestyle='--')
axs[1,0].axvline(24.9, color='orange', linestyle='--')
axs[1,0].axvline(29.9, color='red', linestyle='--')

axs[1,0].axvspan(0, 18.5, color='cyan', alpha=0.15, label='Underweight')
axs[1,0].axvspan(18.5, 24.9, color='green', alpha=0.15, label='Normal')
axs[1,0].axvspan(24.9, 29.9, color='orange', alpha=0.15, label='Overweight')
axs[1,0].axvspan(29.9, max(df3['BMI'].max(), df4['BMI'].max()), color='red', alpha=0.15, label='Obese')
axs[1,0].legend()

#general
sns.kdeplot(df5['BMI'], color='gold', fill=True, label='General', ax=axs[1,1])

axs[1,1].set_title('BMI Distribution - Combined Data')
axs[1,1].set_xlabel('BMI')
axs[1,1].set_ylabel('Density')

axs[1,1].axvline(18.5, color='green', linestyle='--')
axs[1,1].axvline(24.9, color='orange', linestyle='--')
axs[1,1].axvline(29.9, color='red', linestyle='--')

axs[1,1].axvspan(0, 18.5, color='cyan', alpha=0.15, label='Underweight')
axs[1,1].axvspan(18.5, 24.9, color='green', alpha=0.15, label='Normal')
axs[1,1].axvspan(24.9, 29.9, color='orange', alpha=0.15, label='Overweight')
axs[1,1].axvspan(29.9,df5['BMI'].max(), color='red', alpha=0.15, label='Obese')
axs[1,1].legend()
add_bmi_labels(axs[1,1], df5)

plt.tight_layout()
plt.show()

```

```python
# The healthy temperature range for an adult is typically between 97.6F (36.5 C) and 99.6 F (37.5 C).

def add_temp_labels(ax, df, column='Temperature'):

    total = len(df)
    below = ((df[column] < 97.6).sum() / total) * 100
    normal = ((df[column] >= 97.6) & (df[column] <= 99.6)).sum() / total * 100
    above = ((df[column] > 99.6).sum() / total) * 100


    labels = [ f"B:{below:.1f}%",
               f"N:{normal:.1f}%",
               f"A:{above:.1f}%" ]


    positions = [97, 98.5, 100]
    
    for pos, label in zip(positions, labels):
        ax.text(pos, ax.get_ylim()[1] * 0.8, label, fontsize=10,
                ha='center', va='center', bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))
        
fig, axs = plt.subplots(2,2, figsize=(18, 10)) 

#female
sns.kdeplot(df3['Temperature'], color='red', fill=True, ax=axs[0,0])
axs[0,0].set_title('Temperature Distribution - Female')
axs[0,0].set_xlabel('Temperature')
axs[0,0].set_ylabel('Density')

axs[0,0].axvline(97.6, color='green', linestyle='--')
axs[0,0].axvline(99.6, color='orange', linestyle='--')

axs[0,0].axvspan(df3['Temperature'].min(), 97.6, color='cyan', alpha=0.15, label='Below Normal')
axs[0,0].axvspan(97.6, 99.6, color='green', alpha=0.15, label='Normal')
axs[0,0].axvspan(99.6, df3['Temperature'].max(), color='red', alpha=0.15, label='Above Normal')
axs[0,0].legend()
add_temp_labels(axs[0,0], df3)

#male
sns.kdeplot(df4['Temperature'], color='blue', fill=True, ax=axs[0,1])
axs[0,1].set_title('Temperature Distribution - Male')
axs[0,1].set_xlabel('Temperature')
axs[0,1].set_ylabel('Density')

axs[0,1].axvline(97.6, color='green', linestyle='--')
axs[0,1].axvline(99.6, color='orange', linestyle='--')

axs[0,1].axvspan(df4['Temperature'].min(), 97.6, color='cyan', alpha=0.15, label='Below Normal')
axs[0,1].axvspan(97.6, 99.6, color='green', alpha=0.15, label='Normal')
axs[0,1].axvspan(99.6, df4['Temperature'].max(), color='red', alpha=0.15, label='Above Normal')
axs[0,1].legend()
add_temp_labels(axs[0,1], df4)


#combined

sns.kdeplot(df3['Temperature'], color='red', fill=True, ax=axs[1,0],label='Female')
sns.kdeplot(df4['Temperature'], color='blue', fill=True, ax=axs[1,0],label='Male')
axs[1,0].set_title('Temperature Distribution - Male vs Female')
axs[1,0].set_xlabel('Temperature')
axs[1,0].set_ylabel('Density')

axs[1,0].axvline(97.6, color='green', linestyle='--')
axs[1,0].axvline(99.6, color='orange', linestyle='--')

axs[1,0].axvspan(min(df4['Temperature'].min(),df3['Temperature'].min()), 97.6, color='cyan', alpha=0.15, label='Below Normal')
axs[1,0].axvspan(97.6, 99.6, color='green', alpha=0.15, label='Normal')
axs[1,0].axvspan(99.6,max(df4['Temperature'].max(),df3['Temperature'].max()) , color='red', alpha=0.15, label='Above Normal')
axs[1,0].legend()

#general

sns.kdeplot(df5['Temperature'], color='yellow', fill=True, ax=axs[1,1],label='General')
axs[1,1].set_title('Temperature Distribution - Combined Data')
axs[1,1].set_xlabel('Temperature')
axs[1,1].set_ylabel('Density')

axs[1,1].axvline(97.6, color='green', linestyle='--')
axs[1,1].axvline(99.6, color='orange', linestyle='--')

axs[1,1].axvspan(df5['Temperature'].min(), 97.6, color='cyan', alpha=0.15, label='Below Normal')
axs[1,1].axvspan(97.6, 99.6, color='green', alpha=0.15, label='Normal')
axs[1,1].axvspan(99.6,df5['Temperature'].max() , color='red', alpha=0.15, label='Above Normal')
axs[1,1].legend()
add_temp_labels(axs[1,1], df5)

plt.tight_layout()
plt.show()
```

```python
def add_bp_labels(ax, df, column='Blood Pressure'):
    total = len(df)
    healthy = ((df[column] < 120).sum() / total) * 100
    elevated = ((df[column] >= 120) & (df[column] <= 129)).sum() / total * 100
    stage1 = ((df[column] >= 130) & (df[column] <= 139)).sum() / total * 100
    stage2 = ((df[column] >= 140)).sum() / total * 100

    labels = [
        f"H:{healthy:.1f}%",
        f"E:{elevated:.1f}%",
        f"S1:{stage1:.1f}%",
        f"S2:{stage2:.1f}%"
    ]

    positions = [110, 124, 134, 145]
    
    for pos, label in zip(positions, labels):
        ax.text(pos, ax.get_ylim()[1] * 0.8, label, fontsize=10,
                ha='center', va='center',
                bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

        
fig, axs = plt.subplots(2,2, figsize=(18, 10)) 

#female
sns.kdeplot(df3['Blood Pressure'], color='red', fill=True, ax=axs[0,0])
axs[0,0].set_title('Blood Pressure - Female')
axs[0,0].set_xlabel('Blood Pressure')
axs[0,0].set_ylabel('Density')

axs[0,0].axvline(120, color='green', linestyle='--')
axs[0,0].axvline(129, color='orange', linestyle='--')
axs[0,0].axvline(139, color='red', linestyle='--')

axs[0,0].axvspan(df3['Blood Pressure'].min(), 120, color='cyan', alpha=0.15, label='Healthy')
axs[0,0].axvspan(120, 129, color='green', alpha=0.15, label='Elevated')
axs[0,0].axvspan(129, 139, color='orange', alpha=0.15, label='Stage 1 Hypertension')
axs[0,0].axvspan(139, df3['Blood Pressure'].max(), color='red', alpha=0.15, label='Stage 2 Hypertension')
add_bp_labels(axs[0,0], df3)

axs[0,0].legend()


#male
sns.kdeplot(df4['Blood Pressure'], color='blue', fill=True, ax=axs[0,1])
axs[0,1].set_title('Blood Pressure - Male')
axs[0,1].set_xlabel('Blood Pressure')
axs[0,1].set_ylabel('Density')

axs[0,1].axvline(120, color='green', linestyle='--')
axs[0,1].axvline(129, color='orange', linestyle='--')
axs[0,1].axvline(139, color='red', linestyle='--')

axs[0,1].axvspan(df4['Blood Pressure'].min(), 120, color='cyan', alpha=0.15, label='Healthy')
axs[0,1].axvspan(120, 129, color='green', alpha=0.15, label='Elevated')
axs[0,1].axvspan(129, 139, color='orange', alpha=0.15, label='Stage 1 Hypertension')
axs[0,1].axvspan(139, df4['Blood Pressure'].max(), color='red', alpha=0.15, label='Stage 2 Hypertension')

add_bp_labels(axs[0,1], df4)


axs[0,1].legend()

#combined

sns.kdeplot(df3['Blood Pressure'], color='red', fill=True, ax=axs[1,0],label='Female')
sns.kdeplot(df4['Blood Pressure'], color='blue', fill=True, ax=axs[1,0],label='Male')
axs[1,0].set_title('Blood Pressure - Male vs Female')
axs[1,0].set_xlabel('Blood Pressure')
axs[1,0].set_ylabel('Density')

axs[1,0].axvline(120, color='green', linestyle='--')
axs[1,0].axvline(129, color='orange', linestyle='--')
axs[1,0].axvline(139, color='red', linestyle='--')

axs[1,0].axvspan(min(df3['Blood Pressure'].min(), df4['Blood Pressure'].min()), 120, color='cyan', alpha=0.15, label='Healthy')
axs[1,0].axvspan(120, 129, color='green', alpha=0.15, label='Elevated')
axs[1,0].axvspan(129, 139, color='orange', alpha=0.15, label='Stage 1 Hypertension')
axs[1,0].axvspan(139, max(df3['Blood Pressure'].max(), df4['Blood Pressure'].max()), color='red', alpha=0.15, label='Stage 2 Hypertension')

axs[1,0].legend()

#general
sns.kdeplot(df5['Blood Pressure'], color='yellow', fill=True, ax=axs[1,1],label='General')
axs[1,1].set_title('Blood Pressure - Combined Data ')
axs[1,1].set_xlabel('Blood Pressure')
axs[1,1].set_ylabel('Density')

axs[1,1].axvline(120, color='green', linestyle='--')
axs[1,1].axvline(129, color='orange', linestyle='--')
axs[1,1].axvline(139, color='red', linestyle='--')

axs[1,1].axvspan(df5['Blood Pressure'].min(), 120, color='cyan', alpha=0.15, label='Healthy')
axs[1,1].axvspan(120, 129, color='green', alpha=0.15, label='Elevated')
axs[1,1].axvspan(129, 139, color='orange', alpha=0.15, label='Stage 1 Hypertension')
axs[1,1].axvspan(139, df5['Blood Pressure'].max(), color='red', alpha=0.15, label='Stage 2 Hypertension')

add_bp_labels(axs[1,1], df5)

axs[1,1].legend()
plt.show()
```

```python
def add_chol_labels(ax, df, column='Cholesterol'):

    total = len(df)
    healthy = ((df[column] < 200).sum() / total) * 100
    atrisk = ((df[column] >= 200) & (df[column] <= 239)).sum() / total * 100
    high = ((df[column] >= 240)).sum() / total * 100

    labels = [f"H:{healthy:.1f}%",  
              f"R:{atrisk:.1f}%",    
              f"D:{high:.1f}%"     ]

    positions = [180, 215, 250]
    
    for pos, label in zip(positions, labels):
        ax.text(pos, ax.get_ylim()[1] * 0.8, label, fontsize=10,
                ha='center', va='center',
                bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

        
fig, axs = plt.subplots(2,2, figsize=(18, 10)) 

#female
sns.kdeplot(df3['Cholesterol'], color='red', fill=True, ax=axs[0,0])
axs[0,0].set_title('Cholesterol - Female')
axs[0,0].set_xlabel('Cholesterol')
axs[0,0].set_ylabel('Density')

axs[0,0].axvline(200, color='green', linestyle='--')
axs[0,0].axvline(239, color='orange', linestyle='--')

axs[0,0].axvspan(df3['Cholesterol'].min(), 200, color='cyan', alpha=0.15, label='Healthy')
axs[0,0].axvspan(200,239, color='green', alpha=0.15, label='At-Risk')
axs[0,0].axvspan(240, df3['Cholesterol'].max(), color='red', alpha=0.15, label='Dangerous')

add_chol_labels(axs[0,0], df3) 
axs[0,0].legend()


#male
sns.kdeplot(df4['Cholesterol'], color='blue', fill=True, ax=axs[0,1])
axs[0,1].set_title('Cholesterol - Male')
axs[0,1].set_xlabel('Cholesterol')
axs[0,1].set_ylabel('Density')

axs[0,1].axvline(200, color='green', linestyle='--')
axs[0,1].axvline(239, color='orange', linestyle='--')

axs[0,1].axvspan(df4['Cholesterol'].min(), 200, color='cyan', alpha=0.15, label='Healthy')
axs[0,1].axvspan(200,239, color='green', alpha=0.15, label='At-risk')
axs[0,1].axvspan(240, df4['Cholesterol'].max(), color='red', alpha=0.15, label='Dangerous')

add_chol_labels(axs[0,1], df4) 
axs[0,1].legend()

#combined

sns.kdeplot(df3['Cholesterol'], color='red', fill=True, ax=axs[1,0],label='Female')
sns.kdeplot(df4['Cholesterol'], color='blue', fill=True, ax=axs[1,0],label='Male')
axs[1,0].set_title('Cholesterol - combined')
axs[1,0].set_xlabel('Cholesterol')
axs[1,0].set_ylabel('Density')

axs[1,0].axvline(200, color='green', linestyle='--')
axs[1,0].axvline(239, color='orange', linestyle='--')

axs[1,0].axvspan(min(df4['Cholesterol'].min(),df3['Cholesterol'].min()), 200, color='cyan', alpha=0.15, label='Healthy')
axs[1,0].axvspan(200,239, color='green', alpha=0.15, label='At-Risk')
axs[1,0].axvspan(240,max(df4['Cholesterol'].max(),df3['Cholesterol'].max()), color='red', alpha=0.15, label='Dangerous')
axs[1,0].legend()

#general
sns.kdeplot(df5['Cholesterol'], color='yellow', fill=True, ax=axs[1,1],label='General')
axs[1,1].set_title('Cholesterol')
axs[1,1].set_xlabel('Cholesterol')
axs[1,1].set_ylabel('Density')

axs[1,1].axvline(200, color='green', linestyle='--')
axs[1,1].axvline(239, color='orange', linestyle='--')

axs[1,1].axvspan(df5['Cholesterol'].min(), 200, color='cyan', alpha=0.15, label='Healthy')
axs[1,1].axvspan(200,239, color='green', alpha=0.15, label='At-Risk')
axs[1,1].axvspan(240,df5['Cholesterol'].max(), color='red', alpha=0.15, label='Dangerous')

add_chol_labels(axs[1,1], df5) 
axs[1,1].legend()
plt.show()
```
